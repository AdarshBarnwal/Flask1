```python
Q1.What is Flask Framework? What are the advantages of Flask Framework?
Ans:-
     Flask is a micro web framework written in Python. It is classified as a micro framework because it does not require 
     particular tools or libraries. Flask aims to keep the core of an application simple and flexible, while allowing developers
     to easily add extensions and features as needed.

#Advantages of Flask Framework :

1.Lightweight: Flask is a lightweight framework with a small code base, making it easy to learn and use.
2.Flexibility: Flask is highly flexible and allows developers to customize the framework to fit their specific needs.
3.Modular: Flask is modular, which means developers can easily add or remove components as needed.
4.Easy to get started: Flask is easy to get started with and has a minimalistic syntax that makes it accessible to developers of all skill levels.
5.Large community: Flask has a large and active community, which means developers can easily find support and resources online.
6.Good documentation: Flask has good documentation, making it easy for developers to learn and use the framework.
7.Overall, Flask is a great choice for building small to medium-sized web applications, as it is lightweight, flexible, and easy to use.
```

```python
Q2.
Ans:-

#code

from flask import Flask

app = Flask(__name__)

@app.route("/")
def hello_world():
    return "<h1>Hello, World!</h1>"

if __name__=="__main__":
    app.run(host="0.0.0.0")
```

```python
Q4.
Ans:-

#code
from flask import Flask

app = Flask(__name__)

@app.route("/Welcome")
def abc_welcome():
    return "<h1> Welcome to Abc corporation </h1>"

@app.route("/")
def abc_details():
    return "<h2> company name : abc corporation</h2> \n <h2> location : india</h2> \n <h2> contact detail : 999.999.9999 </h2>"

if __name__=="__main__":
    app.run(host="0.0.0.0")
```
